```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('05-fillarea.csv')
```

```python
data.head()
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

### 填滿

```python
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, alpha=0.3)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

#總體
overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, alpha=0.3)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

#總體
overall_median = 57287

#overall_median在y2的參數位置
plt.fill_between(ages, py_salaries, overall_median,
                where=(py_salaries > overall_median), 
                interpolate=True, alpha=0.3)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

#總體
overall_median = 57287

#overall_median在y2的參數位置
plt.fill_between(ages, py_salaries, overall_median,
                where=(py_salaries > overall_median), 
                interpolate=True, alpha=0.3)

plt.fill_between(ages, py_salaries, overall_median,
                where=(py_salaries <= overall_median), 
                interpolate=True, alpha=0.3)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

#總體
overall_median = 57287

#overall_median在y2的參數位置
plt.fill_between(ages, py_salaries, overall_median,
                where=(py_salaries > overall_median), 
                interpolate=True, alpha=0.3)

plt.fill_between(ages, py_salaries, overall_median,
                where=(py_salaries <= overall_median), 
                interpolate=True, color='red', alpha=0.3)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')


#overall_median在y2的參數位置
plt.fill_between(ages, py_salaries, dev_salaries,
                where=(py_salaries > dev_salaries), 
                interpolate=True, alpha=0.3)

plt.fill_between(ages, py_salaries, dev_salaries,
                where=(py_salaries <= dev_salaries), 
                interpolate=True, color='red', alpha=0.3)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
