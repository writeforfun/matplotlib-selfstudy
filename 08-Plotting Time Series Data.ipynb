```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn')
```

```python
dates = [
    datetime(2020, 5, 24),
    datetime(2020, 5, 25),
    datetime(2020, 5, 26),
    datetime(2020, 5, 27),
    datetime(2020, 5, 28),
    datetime(2020, 5, 29),
    datetime(2020, 5, 30)
]
y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y)
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()# get current figure
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate() #get current figure

date_format = mpl_dates.DateFormatter('%b, %d %Y')

plt.gca().xaxis.set_major_formatter(date_format) #get current axis

plt.show()
```

### read csv

```python
data = pd.read_csv('08-timeseries.csv')
data.head()
```

```python
price_date = data['Date']
price_close = data['Close']
```

### 目前的x軸是字串不是datetime,所以要改

```python
plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

### 把字串轉到datetime

```python
data['Date'] = pd.to_datetime(data['Date'])
# data.sort('Date', inplace=True) ##dataframe沒有sort
data.sort_values('Date', inplace=True)
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
